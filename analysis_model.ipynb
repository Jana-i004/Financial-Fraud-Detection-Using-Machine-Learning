```python
from sklearn.model_selection import train_test_split # for splitting data into train test
from sklearn.preprocessing import StandardScaler #scalling data
from sklearn.linear_model import LogisticRegression # modeling side
from sklearn.metrics import classification_report , confusion_matrix # for mode evaluation
from sklearn.pipeline import Pipeline # for training the model  and transformation
from sklearn.compose import ColumnTransformer # for data transformation
from sklearn.preprocessing import OneHotEncoder # for data transformation
```

```python
df.head()
```

```python
df_model = df.drop(["nameOrig" , "nameDest" , "isFlaggedFraud"] , axis = 1)
```

```python
df_model.head()
```

```python
categorical = ["type"]
numeric = ["amount" , "oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]
```

```python
y =  df_model["isFraud"]
X = df_model.drop("isFraud" , axis=1)
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , stratify = y)
```

```python
preprocessor = ColumnTransformer(
    transformers= [
        ("num" , StandardScaler(),numeric) ,
        ("cat", OneHotEncoder(drop="first") , categorical)
    ],
    remainder = "drop"
)
```

```python
pipeline = Pipeline([
    ("prep", preprocessor),
    ("clf",LogisticRegression(class_weight="balanced" , max_iter = 1000))#handling class impalance (99 of data is not fraud so)
])
```

```python
pipeline.fit(X_train , y_train) #training the model
```

```python
y_pred = pipeline.predict(X_test)
```

```python
print(classification_report(y_test ,y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```

```python
pipeline.score(X_test , y_test) *100
```

```python
import joblib
joblib.dump(pipeline,"fraud_detection_pipeline.pkl")
```
